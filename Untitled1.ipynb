```python
#Equação da reta dos perceptrons no plano "cartesiano" a*x + b*y + c = 0 traduzindo pra forma certa sem ser matematica, ficaria w1*x1 + w2*x2 + b = 0
#W = weight e o B(baias) = vies pois ele só vai alterar a reta no espaço mas não mudar a angularidade da reta
#Para a criação da linha, é necessário o A, B e C ou W1, W2, B... Junto deles, o valor aleatório criado para X e depois a formula matematica no Y, e assim plotando os dois valores se cria a linha
#Aprendendo Camada Linear do PyTorch para criar um Perceptron.
#Chamada de função LIMIAR esse "interruptor" que envolve só 0 e 1 sem arredondamento
#Função SIGMOIDE mais usada que a limiar, pois tem uma variedade e um arredondamento dos numeros, onde da mais precisão de valores e sentimentos
#Função tangente perbolica, diferente da sigmoide permite valores negativos nos valores, de resto faz tdo igual
#Função RELU onde basicamente ela é linear para valores positivos, e corta valores negativos, zerando tdo
#Anotarei qual o melhor momento para usar cada uma
#O que é um perceptron: O perceptron é um tipo de rede neural artificial,onde pode se dizer que seria o tipo mais simples de rede neural como um classificador linear.
#X é a variável que guarda os dados da minha destribuição, o Y é o que guarda os rótulos da classe
#Um perceptron é uma reta
#In_features = entrada, sempre é o tamanho da entrada, quantos atributos ou dimensões vc tem. e o Out_features seria a quantidade de neuronios ou perceptrons(linhas) de saida
#Para o problema de XOR, com 2 entradas e 6 saidas, melhor usar o relu, e se tiver 6 entradas e 1 saida, melhor a sigmoide
#O Papel da ativação da função não linear na camada intermediaria (do centro) é extremametne excessencial para atender problemas não lineares se não n consegue solucionar a maioria dos problemas do mundo real
#Deep por causa que é profundo tem várias camadas , e o learning seria mesmo por causa do aprendizado da maquina (aprendizado heirarquico)
#Deep learning é onde se cria uma rede profunda para o aprendizado de caracteristicas para realizar uma inferencia
import numpy as np
import matplotlib.pyplot as plt #Biblioteca de visualização da reta

a = -1
b = 4
c = 0.4 #Se eu mudar o valor de C, vou mudar a reta de lugar, não a angulação

def plotline(a, b, c):
    x = np.linspace(-2, 4, 50) #np.linspace serve para gerar valores para o X, no caso de -2 ate o 4 podendo ter 50 valores
    #dado que temos o valor de X, não fixo mas temos, da para descobrir o valor de Y a partir disso com a função matematica
    #ax + by + c = 0
    #y = (-a*x - c)/b
    y = (-a*x - c)/b


    plt.axvline(0, -1, 1, color='k', linewidth = 3) #Criando a linha do plano cartesiano, vertical e horizontal
    plt.axhline(0, -2, 4, color='k', linewidth = 1) #Criando a linha do plano cartesiano, vertical e horizontal
    plt.plot(x, y) #Função para exibição da linha do grafico
    plt.grid(True)

plotline(a, b, c)


p1 = (2, 0.4) #p1 passando novos valores para X e Y
ret1 = a*p1[0], b*p1[1] + c #Fazendo a prova real e executando a equação com os valores dos pontos em cima da linha
print(ret1) #Teste para ver se quando resolvido a equação, o resultado daria 0, o que no caso deu, com um -2 e o outro 2
p2 = (1, 0.6) #Criando um ponto acima da linha, resultado da equação deve ser positivo
ret2 = a*p2[0], b*p2[1] + c
print(ret2) #valor tem que dar positivo pois o ponto está acima da reta
p3 = (3, -0.4)
ret3 = a*p3[0], b*p3[1] + c
print(ret3) #Valor tem que dar negativo pois está abaixo da reta
#Valores negativos ou positivos, dependem da distancia que está da reta, se estiver muito longe pra cima, vai dar cada vez um valor maior positivo, se estiver mais longe, cada vez mais negativo


plt.plot(p1[0], p1[1], marker='o', color='b')
plt.plot(p2[0], p2[1], marker='o', color='r')
plt.plot(p3[0], p3[1], marker='o', color='g')

#Regrinha, qualquer ponto criado sobre a reta, ao resolver a equação da reta pra aquele ponto, o retorno deve ser 0 ax+by+c = 0
#Se o ponto estiver acima da reta criada, o valor retornado será positivo, e se estiver abaixo em qualquer lugar, será negativo

#Se a entrada tiver 2 dimensões ou mais, ficaria assim
#y = w1*x1 + w2*x2 +b
#y = w1*x1 + w2*x2 + w3*x3 +b
```

```python
from sklearn.datasets import make_classification
import matplotlib.pyplot as plt
import numpy as np

#Aprendendo Camada Linear do PyTorch para criar um Perceptron.

np.random.seed(46) #Travando a raiz de alaetoriazação dos dados com seed 46, assim n vai ficar sendo aleatorio igual antes

X, Y = make_classification(n_features=2, n_redundant=0, n_informative=1, n_clusters_per_class=1)
#X seria as cordenadas, e o Y a classe do ponto, se vale 0 ou 1

#plt.plot(X[:, 0], X[:, 1], marker='o') #[:, 0] seria para mostrar todos os plots de valor 0 e depois de valor 1
#plt.plot faria os pontos todos com linhas entre eles, então neste caso ficaria ruim a visualização, então precisamos usar o scatter
plt.scatter(X[:, 0], X[:, 1], marker='o', c=Y, edgecolor='k') #o C seria um parametro do skatter ali no make_classification para identificar as bolinhas de valores de 0 e de 1, deixando azul e amarelo

p = X[10]
print(Y[10])
plt.plot(p[0], p[1], marker='^', markersize=20)


def plotmodel(w1, w2, b):
    x = np.linspace(-2, 4, 50) #np.linspace serve para gerar valores para o X, no caso de -2 ate o 4 podendo ter 50 valores
    #dado que temos o valor de X, não fixo mas temos, da para descobrir o valor de Y a partir disso com a função matematica
    #ax + by + c = 0
    #y = (-a*x - c)/b
    y = (-w1*x -b)/w2

    xmin, xmax = plt.gca().get_xlim() #Definindo os limites minimo e maximo para fixar o grafico na destribuição
    ymin, ymax = plt.gca().get_ylim()


    plt.scatter(X[:, 0], X[:, 1], marker='o', c=Y, edgecolor='k')

    plt.axvline(0, -1, 1, color='k', linewidth = 1) #Criando a linha do plano cartesiano, vertical e horizontal
    plt.axhline(0, -2, 4, color='k', linewidth = 1) #Criando a linha do plano cartesiano, vertical e horizontal
    plt.plot(x, y) #Função para exibição da linha do grafico
    plt.grid(True)

    plt.xlim(xmin, xmax)
    plt.ylim(ymin, ymax)

w1 = -3
w2 = 5
b = 0
plotmodel(w1, w2, b)


#Parte do codigo para classificação de qualquer ponto novo criado, se ele receberá a cor amarela = 1 ou azul = 0
#Chamada de função LIMIAR esse "interruptor" que envolve só 0 e 1 sem arredondamento
#Função SIGMOIDE mais usada que a limiar, pois tem uma variedade e um arredondamento dos numeros, onde da mais precisão de valores e sentimentos
#Função tangente perbolica, diferente da sigmoide permite valores negativos nos valores, de resto faz tdo igual
#Função RELU onde basicamente ela é linear para valores positivos, e corta valores negativos, zerando tdo
#Anotarei qual o melhor momento para usar cada uma
def classify(ponto, w1, w2, b):
  ret = w1 * ponto[0] + w2 * ponto[1] + b

  if ret >= 1:
    return 1, 'yellow'
  else:
    return 0, 'blue'

p = (2, -1)
classe, cor = classify(p, w1, w2, b)
print(classe, cor)

plotmodel(w1, w2, b)
plt.plot(p[0], p[1], marker='^', color=cor, markersize=39)

#Criação de um metodo para ver quantos pontos de acordo com os pontos que tem dentro da tabela, eu acertei com a minha reta
acertos = 0
for k in range(len(X)):
  categ, cor = classify(X[k], w1, w2, b)
  if categ == Y[k]:
    acertos += 1

print("Acurácia: {0}".format(100*acertos/len(X)))

```

```python
import torch
from torch import nn
from sklearn.datasets import make_classification
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(46) #Travando a raiz de alaetoriazação dos dados com seed 46, assim n vai ficar sendo aleatorio igual antes
X, Y = make_classification(n_features=2, n_redundant=0, n_informative=1, n_clusters_per_class=1)

def plotmodel(w1, w2, b):
    plt.scatter(X[:, 0], X[:, 1], marker='o', c=Y, edgecolor='k')

    xmin, xmax = plt.gca().get_xlim() #Definindo os limites minimo e maximo para fixar o grafico na destribuição
    ymin, ymax = plt.gca().get_ylim()


    x = np.linspace(-2, 4, 50) #np.linspace serve para gerar valores para o X, no caso de -2 ate o 4 podendo ter 50 valores
    #dado que temos o valor de X, não fixo mas temos, da para descobrir o valor de Y a partir disso com a função matematica
    #ax + by + c = 0
    #y = (-a*x - c)/b
    y = (-w1*x -b)/w2

    plt.axvline(0, -1, 1, color='k', linewidth = 1) #Criando a linha do plano cartesiano, vertical e horizontal
    plt.axhline(0, -2, 4, color='k', linewidth = 1) #Criando a linha do plano cartesiano, vertical e horizontal
    plt.plot(x, y) #Função para exibição da linha do grafico
    plt.grid(True)

    plt.xlim(xmin, xmax)
    plt.ylim(ymin, ymax)

w1 = 5
w2 = 1
b = 1.2


perceptron = nn.Linear(2, 1)
sigmoide = nn.Sigmoid()

print(perceptron.weight.data) #Usando um metodo para o valor do Perceptron ser aleatorio, rsultando em 2 valores aleatorios
print(perceptron.bias.data) #Usando um metodo para o valor do Bias(y) ser aleatório, resultando em 1 valor aleatório

perceptron.weight = nn.Parameter(torch.Tensor([[w1, w2]])) #Fixando os 2 valores que queremos usar no perceptron (w1, w2)
perceptron.bias = nn.Parameter(torch.Tensor([b])) #Fixando o valor que queremos usar no Bias

markers = ['^', 'v', '>', '<']
colors = ['r', 'g', 'b', 'grey']

plt.figure(figsize=(8, 6))
plotmodel(w1, w2, b)

for k, idx in enumerate([17, 21, 43, 66]):
  x = torch.Tensor(X[idx])

  ret = perceptron(x)
  act = sigmoide(ret)

  act_limiar = 0 if ret.data < 0 else 1 #Retornando se o valor do retorno.data for 0 retornar o 0, se não retornar o 1

  label = 'ret: {:5.2f}'.format(ret.data.numpy()[0]) + ' limiar: {:4.2f}'.format(act_limiar) + ' act: {:5.2f}'.format(act.data.numpy()[0])
  plt.plot(x[0], x[1], marker=markers[k], color=colors[k], label = label)

plt.legend()
plt.show()


#Para esse tipo de problema, ao decidir qual das 3 funções usar, temos
#A sigmoide e a limiar são as que mais estamos usando nesse caso. o limiar resultando em 1 ou 0 dependendo do lado da zona de decisão.
#A Relu, não é nada adequada pra esse tipo de problema, ela só vai atribuir 0 para quem for negativo, basicamente ela só trunca os negativos para 0 e o resto ela mantem os valores normais
#Usando a função da tangente perbolica, oq vai mudar é que vai permitir valores negativos, e podemos classificar a saida do nosso modelo pelo sinal, se for - é uma classe se for + é outra classe, e perto do 0 está na zona de confusão
```

```python
#Um modo de agrupar as camadas todas em um unico objeto
#Aprendendo com o nn.sequencial

from sklearn.datasets import make_moons
import matplotlib.pyplot as plt
import torch
from torch import nn
from torchsummary import summary #Pacote de visualização que dará mais informações a partir da rede, por exemplo quantos parametros ela tem

input_size = 2 #Varia de acordo com o exemplo, nesse caso, temos x1(0 à 1.5) e x2 de (0 à -1)
output_size = 1 #Varia de acordo com o exemplo, se for no caso de classificação binaria da pra treinar um unico perceptron se for por exemplo mais perto de 1 é uma classe, mais perto de 0 seria outra
#Ou usar os valores crus para ditar, se for maior que 0 de um lado da reta, menor que 0 outro lado da reta
hidden_size = 8 #Neste caso são nós que escolhemos, quanto mais parametros colocarmos aqui, mais capaz a rede vai ser de aprender padrões complexos. usamos 4 de exemplo

X1, Y1 = make_moons(n_samples=300, noise=0.2)
plt.scatter(X1[:, 0], X1[:, 1], marker='o', c =Y1, s=25, edgecolor='k')

#Colocando as 3 camadas minimas para um rede capas de aprender funções mais complexicas além da linear
net = nn.Sequential(nn.Linear(in_features=input_size, out_features=hidden_size), #A primeira seria a cadama escondida (hidden)  #Infeatures seria a dimensionalidade do dado
                    nn.ReLU(), #Ativação não linear (não leva parametros de in_features e out_features)
                    nn.Linear(in_features=hidden_size, out_features=output_size)) #Camada de output (saida)
#Regrinha, o tamanho da saida da primeira camada, será igual ao tamanho da entrada da camada de saída, é um padrão

print(net)
summary(net, input_size=(1, input_size)) #Passa a rede, e o tamanho da entrada como padrão
#No summary mostrará que com parametros na camada escondida, terá 24 parametros, e na camada de saida, terá 9 parametros


#Para fazer um forward com os dados, terá que fazer a seguinte coisa
tensor = torch.from_numpy(X1).float() #Tem que transformar a entrada em um tensor, no caso de arrays numpy tem que chamar o numpy, e se fosse uma lista usaria o Torch.Tensor
#Quando der erros de float64 e precisa de float32 ou float qualquer valor, usar o .float() no final para converter ao valor certo e n dar erro
pred = net(tensor)
print(X1.shape)
print(pred.size())
#Dara 300 amostras com 2 dimensoes, ou seja, essas 2 dimensoes passaram na rede e se transformaram em 300 amostras
#E a predição só terá uma dimensao, que será 1 classe ou outra classe.
#Ou seja, para 300 dados com 2 dimensoes, eu tenho 300 predicoes de um unico perceptron na camada de saida
```

```python
#Um modo de agrupar as camadas todas em um unico objeto
#Aprendendo com o nn.module desta vez

class minhaRede(nn.Module): #Indicando que é uma subclasse da classe modulo nn.Module
   def __init__(self, input_size, hidden_size, output_size): #Colocando os valores para inicializar junto com a classe, mas sem determinar um tamanho fixo deles, para que assim
   #A pessoa que for usar, declarar os tamanhos para determinado exercicio
    #Inicializando a classe superior, a classe mae
      super(minhaRede, self).__init__() #inicializando a classe super

      #Init se define a arquitetura
      self.hidden = nn.Linear(input_size, hidden_size)
      self.relu = nn.ReLU()
      self.output = nn.Linear(hidden_size, output_size)

   def forward(self, X):
    #Forward gerar uma saida a partir do X
    hidden = self.relu(self.hidden(X)) #Recebendo a ativação self.relu com a saida do hidden
    output = self.output(hidden) #Recebendo a ativação self.output com o valor de saida do hidden
    return output

input_size = 2 #Varia de acordo com o exemplo, nesse caso, temos x1(0 à 1.5) e x2 de (0 à -1)
output_size = 1 #Varia de acordo com o exemplo, se for no caso de classificação binaria da pra treinar um unico perceptron se for por exemplo mais perto de 1 é uma classe, mais perto de 0 seria outra
#Ou usar os valores crus para ditar, se for maior que 0 de um lado da reta, menor que 0 outro lado da reta
hidden_size = 8 #Neste caso são nós que escolhemos, quanto mais parametros colocarmos aqui, mais capaz a rede vai ser de aprender padrões complexos. usamos 4 de exemplo

net = minhaRede(input_size, hidden_size, output_size) ##Chamando a função minhaRede juntamente com o metodo init automaticamente, sendo assim, vc tem que passar os valores do init, input_size, output_size e hidden_size


#Fazendo o foward
tensor = torch.from_numpy(X1).float() #Tem que transformar a entrada em um tensor, no caso de arrays numpy tem que chamar o numpy, e se fosse uma lista usaria o Torch.Tensor
#Quando der erros de float64 e precisa de float32 ou float qualquer valor, usar o .float() no final para converter ao valor certo e n dar erro
pred = net(tensor)
print(X1.shape)
print(pred.size())

#Dois metodos padrões pertencente a classe, X é variavel qualquer
```
